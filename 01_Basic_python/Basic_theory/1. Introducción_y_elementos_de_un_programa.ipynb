## Introducción

**¿Quién es un programador (en teoría)?**

Un programador es un profesional que se encarga de escribir, de depurar y de revisar todo el código fuente de un software informático para que lleve a cabo determinadas tareas, o incluso para que mejore, incorporando nuevas funciones para hacer que sea una herramienta más completa y eficiente.

**¿Quién es un programador (en realidad)?**

Un programador es un ser humano con síndrome del impostor que pasa largas horas frente a la pantalla tratando de entender la lógica de lo que escribió un día antes, y es pagado por buscar en Google.

**¿Por qué Python?**

* **Lenguaje de alto nivel:** Python es un lenguaje de programación de alto nivel porque los programadores no necesitan recordar la arquitectura del sistema ni administrar la memoria. Esto lo hace súper amigable para los programadores y es una de las características clave de Python. Dominar Python y todos sus conceptos, paquetes y módulos avanzados puede llevar más tiempo. Sin embargo, aprender la sintaxis básica de Python es muy fácil, en comparación con otros lenguajes populares como C, C++ y Java.


* **Fácil de leer:** El código de Python parece palabras simples en inglés. No se utilizan puntos y comas ni corchetes, y las sangrías definen el bloque de código. Puede saber qué se supone que debe hacer el código con solo mirarlo.


* **Gratis y de código abierto:** Python está desarrollado bajo una licencia de código abierto aprobada por OSI. Por lo tanto, es de uso completamente gratuito, incluso con fines comerciales. No cuesta nada descargar Python, incluirlo en su aplicación o modificarlo libremente.


* **Biblioteca estándar robusta:** Python tiene una extensa biblioteca estándar disponible para que cualquiera la use. Esto significa que los programadores no tienen que escribir su código para cada cosa, a diferencia de otros lenguajes de programación. Hay bibliotecas para la manipulación de imágenes, bases de datos, pruebas unitarias, expresiones y muchas otras funcionalidades.


* **Interpretado:** Cuando se interpreta un lenguaje de programación, significa que el código fuente se ejecuta línea por línea, y no todo a la vez. Los lenguajes de programación como C ++ o Java no se interpretan y, por lo tanto, deben compilarse primero para ejecutarlos. No es necesario compilar Python porque el intérprete lo procesa en tiempo de ejecución.


* **Portátil:** Python es portátil en el sentido de que el mismo código se puede usar en diferentes máquinas. Suponga que escribe un código de Python en una Mac. Si desea ejecutarlo en Windows o Linux más tarde, no tiene que realizar ningún cambio.


* **Orientado a objetos y orientado a procedimientos:** Un lenguaje de programación está orientado a objetos si enfoca el diseño en torno a datos y objetos, en lugar de funciones y lógica. Por el contrario, un lenguaje de programación está orientado a procedimientos si se enfoca más en funciones (código que se puede reutilizar). Python los dos tipos.


* **Extensible:** Se dice que un lenguaje de programación es extensible si puede extenderse a otros lenguajes. El código de Python también se puede escribir en otros lenguajes como C++.


* **Expresivo:** Python necesita usar solo unas pocas líneas de código para realizar tareas complejas. Por ejemplo, para mostrar Hello World, simplemente necesita escribir una línea: imprimir ("Hola mundo"). Otros lenguajes como Java o C tomarían varias líneas para ejecutar esto.


* **Escrito dinámicamente:** Muchos lenguajes de programación necesitan declarar el tipo de variable. Con Python, el tipo de variable se puede decidir durante el tiempo de ejecución. Esto convierte a Python en un lenguaje tipado dinámicamente. Por ejemplo, si tiene que asignar un valor entero 20 a una variable "x", no necesita escribir int x = 20. Solo tiene que escribir x = 15.


* **Simplifique el desarrollo de software complejo:** Python es un lenguaje de propósito general, es decir, se puede utilizar tanto para desarrollar aplicaciones web y de escritorio como para ejecutar análisis estadístico de conjuntos de datos y desplegar modelos de aprendizaje automático en producción. También puede usar las API y unas bibliotecas de visualización de datos para presentar los datos de una manera más atractiva.


* **Otras funcionalidades avanzadas:** Python tiene otras funcionalidades como generadores (utilizados para crear iteradores) y listas de comprensión (utilizadas para crear nuevas listas a partir de otros iterables). Python también tiene una gestión de memoria automática que elimina la necesidad de asignar y liberar memoria manualmente en el código.


##Elementos de un programa

Un programa en Python es un fichero de texto que contiene expresiones del lenguaje Python. Estas expresiones se crean combinando (en el orden correcto) los elementos básicos del lenguaje. Estos elementos son de diferentes tipos:


* **Palabras reservadas**
* **Funciones integradas**
* **Literales**
* **Operadores**
* **Delimitadores**
* **Identificadores**

### Palabras reservadas

Existen una serie de palabras que forman el núcleo del lenguaje, y que nos servirán para escribir nuestros programas.
Estas palabras no podrán ser usadas para nombrar variables o funciones.

### Funciones integradas

Función: bloque de código reutilizable que se puede ejecutar muchas veces dentro de un programa.

Python incluye algunas funciones por defecto, a las cuales podemos hacer referencia sin importar ninguna biblioteca.

### Literales

Los literales son aquellos datos que Python es capaz de manejar, entre los que se encuentran los números (enteros, decimales, complejos, hexadecimal, etc.) y las cadenas de texto.

### Operadores

Los operadores son los caracteres especiales que nos permiten realizar operaciones lógicas y aritméticas.

Los delimitadores son los caracteres que permiten delimitar, separar o representar expresiones.

### Identificadores

Los identificadores son las palabras que se utilizan para nombrar los elementos (variables, funciones, objetos) creados por el usuario.
Están formados por letras, números y/o guión bajo. Se recomienda no utilizar caracteres especiales para evitar complicaciones en un futuro por incompatibilidad de codificaciones.

**Importante:** el primer caracter siempre ha de ser una letra o un guión bajo.
