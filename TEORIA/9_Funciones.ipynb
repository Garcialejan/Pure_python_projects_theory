## Funciones

Una función permite definir un bloque de código reutilizable que se puede ejecutar muchas veces dentro de un programa.
image.png

**¿Qué ventajas me proporciona crear funciones?**

1. **Reutilización de código :** una de las principales ventajas de crear funciones es que puedes definir un conjunto de instrucciones una vez y luego reutilizarlo en múltiples lugares dentro de tu programa. Esto evita la duplicación de código y facilita el mantenimiento.

2. **Abstracción :** las funciones permiten encapsular detalles complejos en un solo lugar y proporcionar una interfaz simple para usar. Los usuarios de la función no necesitan conocer todos los detalles internos, lo que simplifica el uso del código.

3. **Claridad y legibilidad del código :** usar funciones con nombres descriptivos ayuda a que tu código sea más claro y legible. En lugar de tener un programa largo y difícil de entender, puedes dividirlo en funciones con nombres que indiquen claramente su propósito.

4. **Modularidad :** la modularidad es un principio de diseño de software que consiste en dividir un programa en partes más pequeñas y manejables (módulos o funciones). Esto hace que el código sea más fácil de mantener y depurar, ya que cada función se ocupa de una tarea específica.

5. **Facilita la depuración :** si encuentras un error en tu código, es más fácil identificar y solucionar el problema si has dividido tu programa en funciones más pequeñas y autónomas. Puedes aislar y probar cada función individualmente.

6. **Facilita la colaboración :** cuando trabajas en un proyecto con otros programadores, las funciones proporcionan un mecanismo claro para dividir el trabajo y colaborar de manera efectiva. Cada miembro del equipo puede trabajar en funciones específicas sin interferir en el trabajo de los demás.

7. **Reutilización en proyectos futuros :** una vez que haya creado una función útil en Python, puede reutilizarla en proyectos futuros sin tener que volver a escribirla desde cero. Esto ahorra tiempo y esfuerzo.

Para las funciones tenemos algo importante que es el valor de retorno, o el **return**. Muchas veces podemos confundir este comando con el **print**.

El **print()** imprime el valor en la consola, mientras que el **return** devuelve el valor que luego se puede imprimir en la consola, o almacenarlo en una variable para su uso futuro.

````python
def my_func1():
  return 35

x = my_func1()

print(x)
````

````python
def my_func2():
  print(35)

x = my_func2()
print(x)
````

El **return**, además, sirve para detener la ejecución de la función cuando se cumple una determinada condición.

````python
import random
#Una función que el usuario tenga que adivinar un número aleatorio entre 0 y 9.
#Si el usuario escribe SALIR, entonces finalizar el programa.

def adivina_el_numero():

  numero = random.randrange(0,10)
  entrada = int(input("Introduzca un número: "))
  while entrada != numero:
    entrada = input("Introduzca un número: ")
    if entrada == 'SALIR':
      return -1
    entrada = int(entrada)
  return numero

num = adivina_el_numero()

if num == -1:
  print("El jugador se ha rendido y no ha acertado")
else:
  print(f"¡Has acertado! El valor buscado era {num}")
````

Existe la posibilidad de que una función acepte parámetros por defecto. Para ello, en la declaración y creación de la función, pondremos como último parametro el valor que debería tener ese valor en caso de no encontrarse en la llamada.

**Importante:** siempre se ponen al final.

````python
def my_func (a, b, c=5):
  return a+b+c

print(my_func(1.0,1))
print(my_func(1,1,3))
````

````python
def my_func2 (c=5, a, b):
  return a+b+c

print(my_func(a=1, b=1))
````

También existe la posibilidad de que, a priori, no vayamos a saber cuántos parámetros se pueden pasar a la función. En ese caso tenemos el comando *args. Lo que vaya dentro de este parámetro, se almacenará como una tupla.

````python
lista_alumnos = []
def my_func(a,b,c,d,*argss):
  lista_alumnos.append([a,b,c, sum(argss)/len(argss)])
  return sum(argss)/len(argss)

print(my_func("Diego", "Informática", "4º", None,  3,10,6,7,8))
print(my_func("Carlos", "ADE", "1º", "Sí", 7,4,1))
print(my_func("Diego", "Informática", "4º", None,  1,1,1))
print(lista_alumnos)
````

**Encapsulamiento de las variables en las funciones**

En Python, las variables que se encuentran referenciadas dentro de una función son **globales** implícitamente.

En cambio, si a una variable se la asigna un valor en cualquier lugar dentro del cuerpo de una función, se asumirá que es **local** a no ser que explícitamente se la declare como global.

Es decir, Python primero mira si hay una variable local con ese nombre. En caso de que exista, utiliza esa variable. Si no existe, comprueba que haya una variable global (fuera de la función) con ese valor. Si tampoco existe, muestra error.

````python
def global_func():
  print(f"Mi variable es {tipo}")

def local_func():
  tipo = "local"
  print(f"Mi variable es {tipo}")


tipo = "global"
global_func()
local_func()
print(tipo)
````

Por otra parte, todo aquello que se declara dentro de una función, se queda en la función.

````python
def decl_local_var():
  nuevo_nombre = "Nueva variable"

decl_local_var()
print(nuevo_nombre)
````

````python
def doble_saludo():
  print(var*2)


var = "Hola "
doble_saludo()
````

````python
def doble_saludo():
  var = var*2
  print(var)

var = "Hola "
doble_saludo()
print(var)
````

En el caso que queremos modificar una variable global dentro de una función (entorno local), lo que necesitaremos será declarar esa variable como global dentro de la función.

````python
def doble_saludo():
  global var
  var = var*2
  print(var)

var = "Hola "
doble_saludo()
print(var)
````

**Funciones lambda**

Las funciones lambda están diseñadas para operaciones simples, a menudo para hacer transformaciones de datos o filtros de manera rápida.



```
# f = lambda a: a*a (por ejemplo)
```

Las funciones **lambda** van de la mano de dos comandos:
* Map: para ejecutar una función sobre todos los elementos de un objeto iterable.
* Filter: para filtrar los elementos de un objeto iterable



````python
mi_lista = list(range(10))
print(mi_lista)

mi_lista2 = list(map(lambda elemento:elemento*elemento, mi_lista))
print(mi_lista2)
````

````python
mi_lista = list(range(10))

mi_lista3 = list(filter(lambda elemento: elemento%2==0, mi_lista))
print(mi_lista3)
````
