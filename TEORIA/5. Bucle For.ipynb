## Estructuras de control: bucle for

### Bucle For simples

Los bucles son otra herramienta que permite alterar el flujo normal de un programa y repetir una porción de código tantas veces como queramos.

Los bucles *for* son una estructura de control cíclica. Nos permite ejecutar una o varias líneas de código de forma iterativa, teniendo cierto control y conocimiento sobre qué va a pasar en cada iteración. En el ciclo for necesitamos tener un valor inicial y uno final (saber cuando empieza y cuando acaba).

image.png

En pseudocódigo, la sintaxis del bucle for es la siguiente:



```
for variable_control in elemento_iterable(lista, cadena, rango, etc):
  instrucciones_bucle
```



**Características bucle for**
* La variable de control no es necesario definirla antes del bucle.
* El cuerpo del bucle se ejecuta tantas veces como elementos tenga el elemento iterable.
* El elemento iterable puede ser una lista, una cadena, un rango, etc.
* La costumbre más extendida es utilizar la letra i como variable de control, pero se puede utilizar cualquier otro.

````python
print ("Comienzo")
for numero in [1,2,3]:
  print(f"Hola!. EL valor de i es {i}")
print("Final")
````

Si nos encontramos un bucle con un elemento iterable vacío, el bucle no se ejecuta ninguna vez.

````python
print ("Comienzo")
for i in []:
  print(f"La variable de control ahora vale {i}")
print("Final")
````

Para cada iteración, se le va asignando a la variable de control el valor de cada elemento iterable (de manera secuencial).

````python
print ("Comienzo")
for i in [1,2,3]:
  print(f"Hola! El valor de i es {i} y su cuadrado es {i**2}")
print("Final")
````

Los elementos iterables  pueden ser de cualquier tipo, no solo numéricos. Lo importante es que el bucle se repetirá tantas veces como elementos haya para recorrer.

````python
print ("Comienzo")
for i in ["Diego","CEI",3, True]:
  print(f"Hola! El valor de i es {i}")
print("Final")
````

Si quisiéramos recorrer una cadena, también podríamos.

````python
print("Comienzo")
for i in "AMIGO":
  print(f"Dame una {i}")
print("Final")
````

Aunque en los ejemplos anteriores hemos utilizado pequeñas listas para controlar las ejecuciones del bucle, la práctica más extendida es utilizar la función **range()** como elemento iterable, ya que nos permite, de manera sencilla, controlar el flujo y el número de iteraciones que nos gustaría tener en el bucle for.

````python
print(list(range(10)))
print(list(range(3,10)))
print(list(range(3,10,2)))
````

````python
for i in range(3):
  print(f"Hola! El valor de i es {i}")
````

````python
num = int(input("Cuántas veces quieres que te salude: "))
for i in range(num):
  print("Hola!")
````

### Bucle For anidado

Hablamos de bucles anidados cuando un bucle se encuentra en el bloque de instrucciones de otro bloque.

Al bucle que se encuentra dentro del otro se le puede denominar **bucle interno/interior** y al otro **bucle externo/exterior**.

Aunque no es necesario, se recomienda que los nombres de las variables de control de los bucles anidados no coincidan (típicamente i,j,k...).


````python
for i in range(3):
  for j in range(3):
    print(f"i vale {i} y j vale {j}")
````

En el ejemplo anterior, el bucle externo se ejecuta 3 veces, mientras que el interno se ejecuta 3 veces por cada iteración del bucle externo, lo que quiere decir que en total habrá 9 iteraciones (3 veces externo*3 veces interno).


Es importante mantener el orden en la sangría de los bucles, ya que de la sangría depende que Python interprete que un comando pertenezca al bucle externo o interno.

````python
for i in range(3):
  for j in range(3):
    print(f"j vale {j}")
    print(f"i vale {i}")
````

````python
for i in range(3):
  for j in range(3):
    print(f"j vale {j}")
  print(f"i vale {i}")
````

````python
for i in range(3):
  for j in range(3):
    print(f"j vale {j}")
print(f"i vale {i}")
````
