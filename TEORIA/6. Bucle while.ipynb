## Estructuras de control: bucle while

### Bucle while

Un bucle while permite repetir la ejecución de un grupo de instrucciones mientras que se cumpla una condición (es decir, mientras esa condición sea True).

A diferencia del bucle for, en el bucle while será el usuario el que deba llevar el control del momento de salida del bucle. Para ello, será el usuario el que tenga que declarar la variable de control antes del bucle e irla modificando en cada una de las iteraciones de manera manual.

image.png

En pseudocódigo, sería algo como:



```
variable_control

while condicion:
  bloque_instrucciones
  modif variable_control
```
En el bucle while, el método por el que se entra a la función es la evaluación de una condición, lo que deriva en True o False.
* Si el resultado es True, se ejecuta el cuerpo del bucle, se modifica la condición y se vuelve a evaluar. En caso de volver a ser True, se repite el proceso.
* Si el resultado es False, el cuerpo del bucle no se ejecuta y se sigue


````python
i = 1
while i<=3:
  print(i)
  i+=1

print("Final")
````

El ejemplo anterior podría haberse programado con un bucle for, dado que desde el inicio sabíamos que iba a haber 3 iteraciones.

La situación óptima para usar el bucle while es en el que no sabemos cuántas veces vamos a tener que ejecutar el bucle. Por ejemplo:

````python
numero = int(input("Escriba un numero positivo: "))
while numero <=0:
  print("Ha escrito un número negativo. Inténtelo de nuevo")
  numero = int(input("Escriba un numero positivo: "))

print("Gracias")
````

### Bucle while infinito

Una de las circunstancias ante las que tenemos que tener más cuidado es en la de controlar que el bucle tenga un final, dado que podemos estar ante una condición que se cumpla siempre, el bucle no finalice ante nada y entremos en un bucle infinito.

De manera manual podemos forzar la interrupción del programa con el comando **ctrl+C**, aunque debemos intentar que no suceda.

Algunos ejemplos de bucles infinitos:

* El programador ha olvidado modificar la variable de control y la condición siempre se va a cumplir.

````python
i = 1
while i<=10:
  print(i)
````

* El programador ha escrito una condición que se cumplirá siempre.

````python
i = 1
while i > 0:
  print(i)
  i+=1
````

* El programador ha escrito una condición muy específica. **Consejo**: utilizar siempre comparadores del estilo mayor que o menor que en lugar de comparadores de igualdad/desigualdad

````python
i = 1
while i <= 100:
  print(i)
  i+=2
````

### Control de flujo

Existen declaraciones de control para controlar el orden o flujo de ejecución de los ciclos en función de los valores y la lógica.
En Python tenemos 2 principalmente: *break* y *continue*.

**break**

El comando *break* termina el bucle actual y continúa con la ejecución del programa.
image.png


````python
cadena = "Python"
for i in cadena:
  print(i)
  if i == "h":
    print("Encontramos una 'h'")
    break
    print("Aquí no llega")

print("Final")
````

````python
for i in range(10):
  print(i)
  if i== 6:
    break
````

````python
i = 0
while i <=6:
  print(i)
  i+=1
````

````python
x = 5
while True:
    x -= 1
    print(x)
    if x == 0:
        break
print("Fin del bucle")
````

**continue**

Si aparece un *continue* en Python, la iteración finaliza y se pasa a la siguiente iteración del bucle.

image.png

````python
cadena = 'Python'
for i in cadena:
    if i == 'h':
        continue
    print(i)
````

````python
x = 5
while x > 0:
    x -= 1
    if x == 3:
        continue
    print(x)
````
