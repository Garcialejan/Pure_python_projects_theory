## Estructuras de datos

Python nos ofrece una serie de estructuras de datos en las cuales podemos almacenar nuestros datos de manera ordenada y con posibilidad de acceder a ellos de manera lógica. Hay cuatro tipos de estructura:

* **Listas**

> * Datos heterogéneos
* Acepta repetidos
* Elementos mutables
* Accesibles por índice
>

* **Tuplas**

> * Datos heterogéneos
* Acepta repetidos
* Elementos **inmutables**
* Accesibles por índice
>

* **Conjuntos**

> * Datos heterogéneos
* **No admite repetidos**
* Elementos **inmutables**
* **No accesibles por índice**
>

* **Diccionarios**

> * Datos heterogéneos
* Acepta repetidos
* Claves de tipos **inmutables y únicas**
* **Accesibles por clave**

### Conjuntos

Los conjuntos son otra estructura de datos similar a las listas y las tuplas pero más restrictivos. En este caso, los elementos deben ser inmutables, no se almacenarán elementos repetidos y no son accesibles por índice.

```python
set1 = set()
set1 = {0}
set1
```

```python
set1 = {"Diego", 31, True} #Admite datos heterogéneos
```

Al conjunto podemos pasarle valores repetidos, pero solo se quedarán los valores únicos.

```python
datos = [1,1,2,4,6,5,5,4,10]
conj = set(datos)
print(conj)
```

Como funciones específicas de conjuntos tenemos:

```python
my_set = {"C++", "Python", "Java", "SQL"}

my_set.add("C") # Añadir un elemento
print(f"Add: {my_set}")
my_set.update(["Javascript", "React"]) # Añadir varios elementos
print(f"Update: {my_set}")
my_set.remove("C++")  # Elimina un elemento y genera un error si el elemento no existe
print(f"Remove: {my_set}")
my_set.discard("C++") # Elimina un elemento pero no genera error si el elemento no existe.
print(f"Discard: {my_set}")
my_set.clear()      # Elimina todos los elementos de un conjunto
print(f"Clear: {my_set}")
```

Los conjuntos tienen una serie de métodos propios para combinar varios conjuntos entre sí.

#### Ejercicio 2

```python
def encontrar_duplicados(lista):
    elementos_repetidos = {}
    for elemento in lista:
        if lista.count(elemento) > 1:
            elementos_repetidos[elemento] = lista.count(elemento)
    return elementos_repetidos
```

```python
def main():
    lista = [1, 2, 3, 4, 5, 6, 5, 8, 9, 10, 6]
    duplicados = encontrar_duplicados(lista)

    if duplicados:
        print("Los elementos duplicados y su frecuencia son:")
        for elemento, frecuencia in duplicados.items():
            print(f"{elemento}: {frecuencia} veces")
    else:
        print("La lista no tiene elementos duplicados.")
```

```python
if __name__ == "__main__":
    main()
```

### Diccionarios

Un diccionario es una estructura de datos en la que cada posición está representada por un par clave-valor. Se declaran con **{}**, al igual que con los conjuntos.

```python
dic = dict()
dic = {}
```

```python
dic1 = {"Diego": 31, "Pedro": 38, "Antonio": 35}
dic2 = dict([("Diego", 31), ("Pedro", 38), ("Antonio", 35)])
dic3 = dict(Diego=31, Pedro=38, Antonio=35)
```

Los elementos del diccionario no son accesibles por índices si no por claves. Además, el membership se comprueba con las claves.

```python
print(len(dic1))
print(dic1["Diego"])
print("Diego" in dic1)
```

Otras funciones de diccionarios para su manejo:

```python
dic1 = {"Diego": 31, "Pedro": 38, "Antonio": 35}
dic1["Marta"] = 31 # Añadir nuevo valor al diccionario
print(dic1)
dic1["Diego"] = 32
print(f"Modificado: {dic1}")
dic1.pop("Diego") # Eliminar un elemento
print(dic1)
del dic1["Pedro"]
print(dic1)
dic1.clear()
print(dic1)
del dic1
```

```python
dic1 = {"Diego": 31, "Pedro": 38, "Antonio": 35}
print(dic1.keys()) # Accedemos a las claves
print(dic1.values()) # Accedemos a los valores
print(dic1.items()) # Cogemos el diccionario en forma de lista de tuplas
```

Una clave de un diccionario debe ser de un tipo inmutable. Se puede utilizar un entero, flotante, cadena, booleano o tupla. Una **lista** no puede ser clave de un diccionario

```python
my_dict = {1: 34, 3.0:67, "hello": 89, True: 89, (1,2): 45}
```

Si queremos concatenar diccionarios, podemos utilizar la función **update()**, o concatenarlos mediante {* *diccionario1, * *diccionario2}

```python
dict0 = {"Cuenta1": 9.5, "Cuenta2": 8, "Cuenta3": 7.4}
dict1 = {"Cuenta4": 12.3, "Cuenta5": 17, "Cuenta1": 5}

print(dict0)
dict0.update(dict1)
print(dict0)

```

```python
dict0 = {"Cuenta1": 9.5, "Cuenta2": 8, "Cuenta3": 7.4}
dict1 = {"Cuenta4": 12.3, "Cuenta5": 17, "Cuenta1": 5}
dict2 = {"Cuenta6": 1.3, "Cuenta7": 170, "Cuenta1": 6}
dict3 = {**dict2, **dict1,**dict0}
print(dict3)
print(dict0)
print(dict1)
```
