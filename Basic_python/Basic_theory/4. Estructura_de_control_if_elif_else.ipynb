## Estructuras de control: if-elif-else

### Control if

La estructura de control **if** permite que un programa ejecute unas instrucciones cuando se cumple una o varias condiciones.
image.png



Si transformamos este diagrama de flujo a Python, sería algo como:

```python
velocidad = 120
if velocidad > 120:
  print("Multa. La velocidad máxima es 120")

print("Ha pasado usted por un control de velocidad.")
```

Paso a paso, cabe destacar:
* La primera línea de la estructura de control contiene la expresión a evaluar. Como hemos visto anteriormente, estas expresiones devuelven True o False. Si la expresión devuelve True, ejecuta las instrucciones que vienen a continuación.
* Para que Python "sepa" que no hay nada más que evaluar, es necesario acabar la línea con **":"**.
* En el bloque de instrucciones a ejecutar, lo que hará que Python sepa qué instrucciones son las que están incluidas en la condición es la **sangría del texto**. (Cuando utilicemos intérpretes como jupyter notebook o Google Collab, la sangría del texto la hará de manera automática).
* Cuando terminemos de ejecutar el bloque de instrucciones, bastará con eliminar la sangría y volver al nivel anterior.

### Control if - else

La estructura **if-else** permite que un programa ejecute una serie de instrucciones cuando se cumplan unas condiciones, y ejecute otras instrucciones cuando no se cumplan.

image.png

Llevado a código, esto sería algo como:

```python
edad = 18
if edad <= 18.0:
  print("Usted es menor de edad")
else:
  print("Usted es mayor de edad. Puede pasar")
```

Esta estructura de tipo **if-else** puede construirse también concatenando dos estructuras de tipo **if**, pero esto no es recomendable, ya que, si ponemos dos estructuras **if**, obligamos a Python a evaluar ambas. En cambio, con la estructura **if-else**, con que evalúe una ya sabe cuál de las dos tiene que ejecutar.

```python
edad = 19
if edad < 18:
    print("Usted es menor de edad")
if edad >= 18:
  print("Usted es mayor de edad. Puede pasar")
```

**Sentencias anidadas**

Una sentencia **if-else** puede contener a su vez otra condición anidada.

```python
edad = 17
autorizacion = input('Autorización, contesta, SI/NO').upper()


if edad < 18:
  if autorizacion == 'SI':
    print("Usted es menor pero tiene autorización. Puede pasar")
  else:
    print("Usted es menor de edad. Si no viene con autorización no puede pasar.")
else:
  print("Usted es mayor de edad. Puede pasar")
```

### Control if - elif - else

Hasta ahora hemos visto situaciones en las que se podían evaluar 1 o 2 condiciones pero, ¿qué pasa si una situación requiere evaluar más de dos condiciones?

Para ello tenemos la estructura **if - elif - else**.

image.png

Llevado a código Python, esta estructura quedaría como:

```python
edad = 19
if edad < 0:
  print("Error. La edad no puede ser negativa")
elif edad < 18:
  print("Usted es menor de edad")
elif edad > 120:
  print("Enhorabuena. Es usted la persona más longeva del mundo")
else:
  print("Usted es mayor de edad. Puede pasar")
```

#### IF anidados

```python
genero = input("Por favor, introduzca su género (H o M): ")
talla = input("Por favor, introduzca su talla (S, M, L o XL): ")

if genero == 'M':
  if talla == 'L':
    print("No hay stock de esta talla")
  elif talla in ('S', 'XL'):
    print("Quedan pocas unidades. ¡Date prisa!")
  elif talla == 'M':
    print("Artículo disponible")
  else:
    print("Por favor, introduzca una talla correcta (S, M, L o XL)")
elif genero == 'H':
  if talla in ("S", "M", "XL"):
    print(f"La talla {talla} no tiene stock")
  elif talla == 'L':
    print("Artículo disponible")
  else:
    print("Introduzca una talla correcta (S, M, L o XL)")
else:
  print("Introduzca un género aceptado en el sistema (H o M)")
```
